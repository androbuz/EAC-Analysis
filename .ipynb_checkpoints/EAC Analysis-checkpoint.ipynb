# Progress of East African Countries in the Context of Regional Integration

## Introduction

The East African Community (EAC) is a regional intergovernmental organization comprising seven partner states: Burundi, Kenya, Rwanda, South Sudan, Tanzania, Uganda, and the Democratic Republic of Congo (recently joined). According to the <a href="https://www.eac.int/overview-of-eac">EAC website</a>, the mission of the Community is "to widen and deepen economic, political, social and cultural integration in order to improve the quality of life of the people of East Africa through increased competitiveness, value-added production, trade, and investments."

The EAC was established to promote economic integration, political cooperation, and sustainable development. This research aims to analyze whether these goals have been achieved by the member states, particularly Rwanda, Uganda, Kenya, and Tanzania, which are some of the more established members of the EAC.

The relevance of the EAC is currently under scrutiny due to disagreements among member states regarding the M23 conflict in the Democratic Republic of Congo. Rwanda and Uganda are allegedly supporting the M23 rebels, leading to tensions within the Community. As a concerned citizen of one of the EAC member states, I began to question the effectiveness of the EAC and whether it has fulfilled its promises. Although some may speculate about the potential dissolution of the EAC, this research aims to demonstrate that the EAC has contributed positively to regional development, despite the ongoing conflicts.

#### Area of research
The main question I aim to investigate is: how have East African countries performed in key sectors such as agriculture, health, unemployment, and education since joining the EAC?

I will explore whether:
- Are there patterns of convergence or divergence among member states?
- Which areas (e.g., trade, health, education, governance) show the most improvement and which lag behind?
- How does regional integration correlate with these trends?

To address these questions, I will focus on the following key indicators:
- `Agriculture`: Crop production, agricultural export volumes, and food security metrics.
- `Health`: Access to healthcare services, life expectancy, and health expenditure.
- `Unemployment`: Unemployment rates, job creation trends, and youth employment statistics.
- `Education`: Literacy rates, school enrollment ratios, and educational attainment levels.

These indicators are chosen to reflect the goals of the EAC and provide a comprehensive view of the socio-economic progress made by member countries. The analysis will highlight both the successes and challenges in achieving the EAC’s mission of improving the quality of life for its people through regional cooperation and integration.

- The required data for this project is hosted on the World Bank server and is available in CSV, Excel, or XML formats. <a href="https://data.worldbank.org/country/kenya">Example of the Kenya data</a>.
- To streamline the process, I used functions to programmatically download the data for each of the four countries, allowing for code reuse.
- I chose to download the Excel files, as they include a metadata sheet with detailed explanations of each indicator.

```python
import requests

def download_country_data(country_code):
    #url format for each download link
    url = f"https://api.worldbank.org/v2/en/country/{country_code}?downloadformat=excel"
    file = f"{country_code}_DATA.xls" #create its excel file
    response = requests.get(url) #response from the server
    try:
        with open(file, "wb") as f:
            f.write(response.content) #saving the response content, i.e. excel data
        print(f"data successfully downloaded for {country_code} into {file}")
        return file
    except Exception as e:
        print(f"server error or download failed with error: {e}")
    
```

- I must also convert the .xls file downloaded into a dataframe. 
- Upon analysis, I discovered that the Excel files have the data's title and the year it was recorded; this title spans three rows. The title is not important to my analysis, so I will remove it from the function. 
- The Excel file contains three sheets: the data, metadata, and the origin or source of the data. Since I only need the data, I will return a Pandas data frame of the first sheet.

```python
import xlrd #useful library for excel files manipulation

def excel_to_df(country_code):
    try:
        file = download_country_data(country_code)
        if file is not None:
            #the excel files contain a title spanning 3 rows
            #it also contains 3 sheets, the first being the one with the data
            df = pd.read_excel(file, skiprows=3, sheet_name=0)
            return df
        else:
            print("error creating the file") #can not create the file
            return None
    except Exception as e:
        print(f"error {e} encountered")
```

- Due to the unnatural structure of the data, I need to make a function that gets a particular indicator and a year, and print out a scalar value. Otherwise, the code to query values can be repetitive and lengthy.

```python
def query(df, ind, yr):
    """df, ind as string, yr as number/string"""
    year = str(yr) #making sure the year is a string
    row = df[df["Indicator Name"]==ind]
    value = row[year].values[0]
    return value
```

```python
query(df_RW, "Urban population growth (annual %)", "1961")
```

- Since the data has many null values for a particular year, the other issue will be comparing values of one country with another one which has a NaN value in that particular year.
- For example, Rwanda doesn't have 'an urban population growth recorded for 1960':

```python
query(df_RW, "Urban population growth (annual %)", 1960)
```

- I will also need to query values within a range of years with a starting year to an end year; This will be accomplished by the function below.

```python
def query_range(df, ind, start_yr, end_yr):
    """df, ind as string, yr as number/string"""
    values = [] ##the return array containing values
    years = [] #the return array containing years
    try:
        start_year = int(start_yr) #making sure the year is an integer
        end_year = int(end_yr) #making sure the year is an integer
    except:
        print("error converting to string")
    yrs = range(start_year, end_year)
    for year in yrs:
        year = str(year) #making it back to a string
        row = df[df["Indicator Name"]==ind]
        if not row.empty and year in row.columns:
            value = row[year].values[0]
            values.append(value) #push the values in an array
            years.append(year)
        else:
            values.append(None) #put none
            years.append(None) #put none
    return years, values
```

- To handle the indicators that have missing data for some years, I will need to clean out those NaN values as they can raise exception errors and make the code crash abruptly. That functionality will be handled by the following function:

```python
def removeNan(years, values):
    """Removing null values; takes years, values arrays"""
    cleaned_values = []
    cleaned_years = []
    # extract only non null rows for the tuple year, value
    for i in range(len(values)): 
        if not np.isnan(values[i]):
            #storing the values into respective arrays
            cleaned_values.append(values[i])
            cleaned_years.append(years[i])
    return cleaned_years,cleaned_values
```

- Many datasets contain null values, which can lead to inaccurate analysis. For example, to accurately compare the average male vs. female primary completion rates, I need rows with data for both indicators.This will be achieved using the following function:

```python
def getCommon(years1, years2, values1, values2):
    """Get the common rows of two collections of data of years-values"""
    yrs1, yrs2, vals1, vals2 = [], [], [], []
    set_years2 = set(years2)
    #get equivalent year in the other data
    for i in range(len(years1)):
        for j in range(len(years2)):
            if years1[i] == years2[j]:
                yrs1.append(years1[i]) #update the yrs1 array
                vals1.append(values1[i]) #update the vals1 array
                yrs2.append(years2[j]) #update the yrs1 array
                vals2.append(values2[j]) #update the vals1 array
                break #move to the next i iteration
    return yrs1, yrs2, vals1, vals2
```

Life expectancy at birth indicates the number of years a newborn infant would live if prevailing patterns of mortality at the time of its birth were to stay the same throughout its life.

## Conclusion
The East African Community (EAC) has made notable strides in fostering regional integration and development since the member countries joined. This research examined key areas such as education, health, unemployment, agriculture, and self-employment, providing insights into the collective progress and individual challenges faced by Rwanda, Uganda, Kenya, and Tanzania.

#### Education:
The EAC countries have shown a strong commitment to education, with substantial investments across all levels. Kenya excels in primary education, while Rwanda leads in secondary education. The remarkable spending on tertiary education, particularly by Tanzania, underscores the region's focus on higher education. However, gender disparities remain, with men slightly more represented than women in tertiary education. Rwanda's balanced gender representation at lower education levels highlights its effective policies.

#### Health and Mortality:
Health outcomes, such as life expectancy and mortality rates, vary across the region. Kenya's advancements in healthcare infrastructure have positively impacted its mortality rates, while Rwanda continues to grapple with the legacy of the 1994 genocide. Despite these challenges, all countries are making continuous improvements.

#### Unemployment and Labor Force Participation:
Unemployment remains a pressing issue, particularly for women. Rwanda faces high and growing unemployment rates, which is a concerning trend. Youth labor force participation is strong in Tanzania and Uganda, showcasing the potential of young people in the labor market, though Tanzania's participation is declining over time.

#### Self-employment and Agricultural Employment:
Self-employment reflects the EAC's emphasis on self-reliance. Tanzania leads in this area, with Uganda following closely. Rwanda's decline in self-employment rates points to changing labor dynamics. In agriculture, Rwanda, Uganda, and Tanzania excel in job creation, while Kenya struggles to keep pace.

#### Overall Assessment:
Despite the challenges and disparities, the EAC has proven to be a fruitful initiative. It has fostered regional cooperation, economic growth, and social development. The continuous improvement in key sectors demonstrates the potential for further integration and collective progress. While there are areas needing attention, such as unemployment and gender disparities, the overall impact of the EAC has been positive. Continued commitment to its goals can lead to even greater achievements in the future.

In conclusion, the EAC has played a crucial role in shaping the socio-economic landscape of its member states, and its continued efforts are essential for sustained regional development.
